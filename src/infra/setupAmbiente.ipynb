```python
from pyspark.sql import SparkSession

spark = SparkSession.getActiveSession()

VOLUME_CATALOG = "main"
VOLUME_SCHEMA = "engenharia_dados"
VOLUME_NAME = "aviacao_landing"

META_SCHEMA = "aviacao_meta"
LANDING_SCHEMA = "aviacao_landing"
BRONZE_SCHEMA = "aviacao_bronze"

print(f"Criando catálogo (se não existir): {VOLUME_CATALOG}")
spark.sql(f"CREATE CATALOG IF NOT EXISTS {VOLUME_CATALOG}")

print(f"Criando schema do volume (se não existir): {VOLUME_CATALOG}.{VOLUME_SCHEMA}")
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {VOLUME_CATALOG}.{VOLUME_SCHEMA}")

print(f"Criando volume (se não existir): {VOLUME_CATALOG}.{VOLUME_SCHEMA}.{VOLUME_NAME}")
spark.sql(f"CREATE VOLUME IF NOT EXISTS {VOLUME_CATALOG}.{VOLUME_SCHEMA}.{VOLUME_NAME}")

print(f"Criando schema de metadados (se não existir): {META_SCHEMA}")
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {META_SCHEMA}")

print(f"Criando schema de landing Delta (se não existir): {LANDING_SCHEMA}")
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {LANDING_SCHEMA}")

print(f"Criando schema de bronze (se não existir): {BRONZE_SCHEMA}")
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {BRONZE_SCHEMA}")

landing_path = f"/Volumes/{VOLUME_CATALOG}/{VOLUME_SCHEMA}/{VOLUME_NAME}/aviacao/landing"
print("\nAmbiente criado/ajustado com sucesso.")
print(f"Use este caminho no LANDING_CSV_BASE_PATH:\n{landing_path}")
```
